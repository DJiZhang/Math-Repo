```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import matplotlib.animation as animation
from Fourier import NumericFourier, SymbolicFourier, Animate
```

```python
# Defining a function
x_sym = sp.symbols("x", real=True)
f =  sp.Piecewise((0, x_sym < 0), (1, x_sym >= 0))

# Instantiate from Numeric or Symbolic class
fourier_num = NumericFourier(f, x_sym, 0, 1, half="sin")
fourier_sym = SymbolicFourier(f, x_sym, 0, 1, half="cos")
```

```python

# Basic Plotting
fig, axes = plt.subplots(2,1)
sample = 3000
N = 300
fourier_num.plot_series(axes[0], N, show_f=False, sample=10000, x_lim=2, label="Numerical", linewidth=1)
fourier_sym.plot_series(axes[1], N, show_f=False, sample=10000, x_lim=2, label="Symbolic")
axes[0].legend()
axes[1].legend()
```

```python
# Animations - Run in python file
x_sym = sp.symbols("x", real=True)
f = 1
ani_num = Animate(fourier_num)
fig, axes = plt.subplots(2, 1)
ani = ani_num.heat(fig, axes[0], axes[1], N=N,t1=1, sample=3000)
plt.show()
```
